```python
import sqlite3

# Veritabanına bağlan
connection = sqlite3.connect("yeliz.deliveroo.db")
cursor = connection.cursor()

# Fiyat dağılımını hesaplayan SQL sorgusu
cursor.execute("""                              
SELECT 
    CASE
        WHEN price_float BETWEEN 1.0 AND 10.0 THEN '1.0-10.0'
        WHEN price_float > 10.0 AND price_float <= 20.0 THEN '10.0-20.0'
        WHEN price_float > 20.0 AND price_float <= 30.0 THEN '20.0-30.0'
        WHEN price_float > 30.0 AND price_float <= 50.0 THEN '30.0-50.0'
        ELSE '50.0+'
    END AS price_range,
    COUNT(*) AS item_count
FROM menu_items
WHERE price_float > 1.0
GROUP BY price_range
ORDER BY price_range;
""")

# Sorgunun sonuçlarını al
price_distribution = cursor.fetchall()

# Sonuçları yazdır
print("Price Distribution of Menu Items:")
for price_range, item_count in price_distribution:
    print(f"Price Range: {price_range}, Item Count: {item_count}")

# Bağlantıyı kapat
connection.close()
```

```python
# Deliveroo Database

import sqlite3
import matplotlib.pyplot as plt

# Connect to the database
connection = sqlite3.connect("yeliz.deliveroo.db")
cursor = connection.cursor()

# SQL query to calculate price distribution
cursor.execute("""                              
SELECT 
    CASE
        WHEN price_float BETWEEN 1.0 AND 10.0 THEN '1.0-10.0'
        WHEN price_float > 10.0 AND price_float <= 20.0 THEN '10.0-20.0'
        WHEN price_float > 20.0 AND price_float <= 30.0 THEN '20.0-30.0'
        WHEN price_float > 30.0 AND price_float <= 50.0 THEN '30.0-50.0'
        ELSE '50.0+'
    END AS price_range,
    COUNT(*) AS item_count
FROM menu_items
WHERE price_float > 1.0
GROUP BY price_range
ORDER BY price_range;
""")

# Fetch results
price_distribution = cursor.fetchall()

# Prepare data for plotting
price_ranges = [x[0] for x in price_distribution]  # Get the price range labels
item_counts = [x[1] for x in price_distribution]  # Get the counts for each price range

# Create a bar plot (histogram)
plt.bar(price_ranges, item_counts, color='skyblue')
plt.title('Price Distribution of Menu Items')
plt.xlabel('Price Range')
plt.ylabel('Number of Items')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to make room for the labels

# Show the plot
plt.show()

# Close the connection
connection.close()
```

Question: Identify the World Hummus Order (WHO); top 3 hummus serving restaurants.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to the database
connection = sqlite3.connect("yeliz.deliveroo.db")
cursor = connection.cursor()

# Step 2: Execute the SQL query
cursor.execute("""
SELECT 
    restaurants.name AS restaurant_name,
    COUNT(menu_items.name) AS hummus_item_count
FROM 
    menu_items
JOIN 
    restaurants
ON 
    menu_items.restaurant_id = restaurants.id  
WHERE 
    menu_items.name LIKE '%hummus%'
GROUP BY 
    restaurants.id , restaurants.name , restaurants.rating
ORDER BY 
    restaurants.rating DESC, 
    hummus_item_count DESC
LIMIT 3;
""")

# Fetch results and store in a DataFrame
data = cursor.fetchall()
columns = ['restaurant_name', 'hummus_item_count']
df = pd.DataFrame(data, columns=columns)

# Step 3: Plot the Pie Chart
plt.figure(figsize=(5, 4))
plt.pie(df['hummus_item_count'], labels=df['restaurant_name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 3 Hummus Serving Restaurants')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()

# Close the connection
connection.close()
```

```python
# Deliveroo Database
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the database
connection = sqlite3.connect("yeliz.deliveroo.db")
cursor = connection.cursor()

# Execute the SQL query
cursor.execute("""
SELECT 
    restaurants.name AS restaurant_name,
    COUNT(menu_items.name) AS hummus_item_count,
    AVG(restaurants.rating) AS avg_rating
FROM 
    menu_items
JOIN 
    restaurants
ON 
    menu_items.restaurant_id = restaurants.id  
WHERE 
    menu_items.name LIKE '%hummus%'
GROUP BY 
    restaurants.id , restaurants.name , restaurants.rating
ORDER BY 
    restaurants.rating DESC, 
    hummus_item_count DESC
LIMIT 3;
""")

# Fetch the results and create a DataFrame
data = cursor.fetchall()
columns = ['restaurant_name', 'hummus_item_count', 'avg_rating']  # Ensure the order of columns matches the query
df = pd.DataFrame(data, columns=columns)

# Close the connection
connection.close()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Set bar width
bar_width = 0.35
index = range(len(df))

# Bar chart for hummus item count
bars1 = ax.bar(index, df['hummus_item_count'], bar_width, label='Hummus Item Count', color='tab:orange', alpha=0.7)

# Bar chart for average rating
bars2 = ax.bar([i + bar_width for i in index], df['avg_rating'], bar_width, label='Average Rating', color='tab:blue', alpha=0.7)

# Adding labels and title
ax.set_xlabel('Restaurant Name')
ax.set_ylabel('Count / Rating')
ax.set_title('Top 3 Restaurants: Hummus Item Count and Average Rating')
ax.set_xticks([i + bar_width / 2 for i in index])  # Centering the x-ticks
ax.set_xticklabels(df['restaurant_name'], rotation=45, ha='right')
ax.legend()

# Adding data labels on top of the bars
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.0, yval, ha='center', va='bottom')  # Adding hummus item count labels

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.0, f"{yval:.2f}", ha='center', va='bottom')  # Adding avg rating labels

# Show the plot
plt.tight_layout()  
plt.show()
```

```python
# Takeaway Database
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the database
connection = sqlite3.connect("yeliz.takeaway.db")
cursor = connection.cursor()

# Execute the SQL query
cursor.execute("""
SELECT 
    restaurants.name AS restaurant_name,  
    AVG(restaurants.ratings) AS avg_rating,                   
    COUNT(menuItems.name) AS hummus_item_count 
FROM 
    menuItems                             
JOIN 
    restaurants                           
ON 
    menuItems.primarySlug = restaurants.primarySlug  
WHERE 
    menuItems.name LIKE '%hummus%'        
GROUP BY 
    restaurants.primarySlug, restaurants.name, restaurants.ratings 
ORDER BY 
    restaurants.ratings DESC,              
    hummus_item_count DESC                
LIMIT 3; 
""")

# Fetch the results and create a DataFrame
data = cursor.fetchall()
columns = ['restaurant_name', 'hummus_item_count', 'avg_rating']  # Ensure the order of columns matches the query
df = pd.DataFrame(data, columns=columns)

# Close the connection
connection.close()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Set bar width
bar_width = 0.35
index = range(len(df))

# Bar chart for hummus item count
bars1 = ax.bar(index, df['hummus_item_count'], bar_width, label='Hummus Item Count', color='tab:orange', alpha=0.7)

# Bar chart for average rating
bars2 = ax.bar([i + bar_width for i in index], df['avg_rating'], bar_width, label='Average Rating', color='tab:blue', alpha=0.7)

# Adding labels and title
ax.set_xlabel('Restaurant Name')
ax.set_ylabel('Count / Rating')
ax.set_title('Top 3 Restaurants: Hummus Item Count and Rating')
ax.set_xticks([i + bar_width / 2 for i in index])  # Centering the x-ticks
ax.set_xticklabels(df['restaurant_name'], rotation=45, ha='right')
ax.legend()

# Adding data labels on top of the bars
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.0, yval, ha='center', va='bottom')  # Adding hummus item count labels

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f"{yval:.2f}", ha='center', va='bottom')  # Adding avg rating labels

# Show the plot
plt.tight_layout()  
plt.show()
```
