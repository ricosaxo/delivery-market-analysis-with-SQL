```python
import sqlite3

# Veritabanına bağlan
connection = sqlite3.connect("yeliz.deliveroo.db")
cursor = connection.cursor()

# Fiyat dağılımını hesaplayan SQL sorgusu
cursor.execute("""                              
SELECT 
    CASE
        WHEN price_float BETWEEN 1.0 AND 10.0 THEN '1.0-10.0'
        WHEN price_float > 10.0 AND price_float <= 20.0 THEN '10.0-20.0'
        WHEN price_float > 20.0 AND price_float <= 30.0 THEN '20.0-30.0'
        WHEN price_float > 30.0 AND price_float <= 50.0 THEN '30.0-50.0'
        ELSE '50.0+'
    END AS price_range,
    COUNT(*) AS item_count
FROM menu_items
WHERE price_float > 1.0
GROUP BY price_range
ORDER BY price_range;
""")

# Sorgunun sonuçlarını al
price_distribution = cursor.fetchall()

# Sonuçları yazdır
print("Price Distribution of Menu Items:")
for price_range, item_count in price_distribution:
    print(f"Price Range: {price_range}, Item Count: {item_count}")

# Bağlantıyı kapat
connection.close()
```

Question: Identify the World Hummus Order (WHO); top 3 hummus serving restaurants.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to the database
connection = sqlite3.connect("yeliz.deliveroo.db")
cursor = connection.cursor()

# Step 2: Execute the SQL query
cursor.execute("""
SELECT 
    restaurants.name AS restaurant_name,
    COUNT(menu_items.name) AS hummus_item_count,
    AVG(restaurants.rating) AS avg_rating  -- Include average rating in the query
FROM 
    menu_items
JOIN 
    restaurants
ON 
    menu_items.restaurant_id = restaurants.id  
WHERE 
    menu_items.name LIKE '%hummus%'
GROUP BY 
    restaurants.id , restaurants.name
ORDER BY 
    avg_rating DESC,                    -- Sort by average rating
    hummus_item_count DESC
LIMIT 3;
""")

# Fetch results and store in a DataFrame
data = cursor.fetchall()
columns = ['restaurant_name', 'hummus_item_count', 'avg_rating']  # Add avg_rating to the columns
df = pd.DataFrame(data, columns=columns)

# Step 3: Create custom labels for the Pie Chart
labels = [f"{row[0]} (Rating: {row[2]:.1f})" for row in data]  # Create labels with restaurant name and rating

# Step 4: Plot the Pie Chart
plt.figure(figsize=(5, 4))
plt.pie(df['hummus_item_count'], labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 3 Hummus Serving Restaurants')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()

# Close the connection
connection.close()

```

```python
# Deliveroo Database
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the database
connection = sqlite3.connect("yeliz.deliveroo.db")
cursor = connection.cursor()

# Execute the SQL query
cursor.execute("""
SELECT 
    restaurants.name AS restaurant_name,
    COUNT(menu_items.name) AS hummus_item_count,
    AVG(restaurants.rating) AS avg_rating
FROM 
    menu_items
JOIN 
    restaurants
ON 
    menu_items.restaurant_id = restaurants.id  
WHERE 
    menu_items.name LIKE '%hummus%'
GROUP BY 
    restaurants.id , restaurants.name , restaurants.rating
ORDER BY 
    restaurants.rating DESC, 
    hummus_item_count DESC
LIMIT 3;
""")

# Fetch the results and create a DataFrame
data = cursor.fetchall()
columns = ['restaurant_name', 'hummus_item_count', 'avg_rating']  # Ensure the order of columns matches the query
df = pd.DataFrame(data, columns=columns)

# Close the connection
connection.close()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Set bar width
bar_width = 0.35
index = range(len(df))

# Bar chart for hummus item count
bars1 = ax.bar(index, df['hummus_item_count'], bar_width, label='Hummus Item Count', color='tab:orange', alpha=0.7)

# Bar chart for average rating
bars2 = ax.bar([i + bar_width for i in index], df['avg_rating'], bar_width, label='Average Rating', color='tab:blue', alpha=0.7)

# Adding labels and title
ax.set_xlabel('Restaurant Name')
ax.set_ylabel('Count / Rating')
ax.set_title('Top 3 Restaurants: Hummus Item Count and Average Rating')
ax.set_xticks([i + bar_width / 2 for i in index])  # Centering the x-ticks
ax.set_xticklabels(df['restaurant_name'], rotation=45, ha='right')
ax.legend()

# Adding data labels on top of the bars
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.0, yval, ha='center', va='bottom')  # Adding hummus item count labels

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.0, f"{yval:.2f}", ha='center', va='bottom')  # Adding avg rating labels

# Show the plot
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
import sqlite3
import folium
from folium.plugins import HeatMap

# SQLite veritabanına bağlan
conn = sqlite3.connect("../databases/deliveroo.db")  # Veritabanı dosyanızın yolunu belirtin

# Sorguyu çalıştır ve sonucu bir dataframe'e al
query = """
SELECT DISTINCT restaurants.id, restaurants.latitude, restaurants.longitude, restaurants.postal_code
FROM restaurants
JOIN menu_items ON restaurants.id = menu_items .restaurant_id
WHERE LOWER(menu_items.name) LIKE '%halal%';
"""
data = pd.read_sql_query(query, conn)

# Bağlantıyı kapat
conn.close()



# Harita oluştur
belgium_map = folium.Map(location=[50.8503, 4.3517], zoom_start=8)  # Belçika'nın merkez koordinatları

# Restoranları haritaya ekle
for _, row in data.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Posta Kodu: {row['postal_code']}, Restoran ID: {row['id']}",
        icon=folium.Icon(color='green', icon='cutlery', prefix='fa')
    ).add_to(belgium_map)

# Haritayı kaydet
belgium_map.save("belgium_halal_restaurants.html")



# HeatMap için restoran konumlarını hazırlayın
locations = data[['latitude', 'longitude']].values.tolist()

# Harita oluştur ve HeatMap ekle
belgium_map = folium.Map(location=[50.8503, 4.3517], zoom_start=8)
HeatMap(locations).add_to(belgium_map)

# Haritayı kaydet
belgium_map.save("belgium_halal_heatmap.html")

```

```python
import pandas as pd
import sqlite3
import folium

# Connect to SQLite database
conn1 = sqlite3.connect("../databases/deliveroo.db")  # Veritabanı dosyanızın yolunu belirtin

#  run the query and put it in a data frame 
query1 = """
SELECT DISTINCT restaurants.id, restaurants.latitude, restaurants.longitude, restaurants.postal_code
FROM restaurants
JOIN menu_items ON restaurants.id = menu_items.restaurant_id
WHERE LOWER(menu_items.name) LIKE '%halal%';
"""
data1 = pd.read_sql_query(query1, conn1)

conn1.close()

# connect to second database 
conn2 = sqlite3.connect("../databases/ubereats.db")  
query2 = """
SELECT DISTINCT 
    restaurants.id, 
    restaurants.location__latitude AS latitude,  -- Renaming for clarity
    restaurants.location__longitude AS longitude   -- Renaming for clarity
FROM 
    restaurants
JOIN 
    menu_items ON restaurants.id = menu_items.restaurant_id
WHERE 
    LOWER(menu_items.name) LIKE '%halal%';
"""
data2 = pd.read_sql_query(query2, conn2)
conn2.close() 


# connect to the 3. database
conn3 = sqlite3.connect("../databases/takeaway.db")  # İkinci veritabanı
query3 = """
SELECT DISTINCT 
    restaurants.primarySlug, 
    restaurants.latitude, 
    restaurants.longitude
FROM 
    restaurants
JOIN 
    menuItems ON restaurants.primarySlug = menuItems.primarySlug
WHERE 
    LOWER(menuItems.name) LIKE '%halal%';
"""
data3 = pd.read_sql_query(query3, conn3)
conn3.close() 


# create a map
belgium_map = folium.Map(location=[50.8503, 4.3517], zoom_start=8)  # Belçika'nın merkez koordinatları

# Put the halal restaurants which are in the deliveroo database
for _, row in data1.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Posta Kodu: {row['postal_code']}, Restoran ID: {row['id']}",
        icon=folium.Icon(color='blue', icon='cutlery', prefix='fa')
    ).add_to(belgium_map)

# Put the halal restaurants which are in the ubereats database
for _, row in data2.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],  # Doğru isimler kullanıldı
        popup=f"Restoran ID: {row['id']}",  # Postal kodu kaldırıldı, eğer yoksa hata verir
        icon=folium.Icon(color='green', icon='cutlery', prefix='fa')  # Farklı renk ile ayırt et
    ).add_to(belgium_map)

# Put the halal restaurants which are in the takeaway database
for _, row in data3.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],  # Doğru isimler kullanıldı
        popup=f"primarySlug: {row['primarySlug']}",  # Postal kodu kaldırıldı, eğer yoksa hata verir
        icon=folium.Icon(color='orange', icon='cutlery', prefix='fa')  # Farklı renk ile ayırt et
    ).add_to(belgium_map)

# save the map
belgium_map.save("bbelgium_halal_restaurants.html")

print("Harita oluşturuldu: bbelgium_halal_restaurants.html")
```
